```python
r = {}
r["e"] = x / (y + z)
r["f"] = (x / y) + z
r["g"] = 2 * x / 3 * y
r["h"] = 2 * x / (3 * y)
r["i"] = x * y % z
r["j"] = x * (y % z)
r["k"] = 3 * x - z - 2 * x
r["l"] = 2 * x / 5 % y
r["m"] = x - 100 % y % z
r["n"] = x - y - z * 2
for result in r:
    print(f"resultado {result}): {r[result]}")
```
